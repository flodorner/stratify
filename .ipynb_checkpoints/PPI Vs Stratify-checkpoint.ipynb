```python
import numpy as np
from matplotlib import pyplot as plt
```

```python

def ppi_k(b,p,q, n=100, k=1000):
    
    y = np.random.binomial(1,b,size=(n,k))
    f = np.random.binomial(1, y*p + (1-y) * (1-q) )

    sum_y_l  = (y).sum(axis=0)
    sum_f_l  = (f).sum(axis=0)
    sum_yf_l = (y * f).sum(axis=0)

    mean_y_l = sum_y_l / n
    mean_f_l = sum_f_l / n
    
    mean_f_u = b*p + (1-b) * (1-q)
    var_f_u = mean_f_u * (1-mean_f_u)

    # Covariance on labeled set: matches np.cov(yl, fl)[0,1] default ddof=1
    cov_yf_l = (sum_yf_l - n * mean_y_l * mean_f_l) / (n - 1)
    denom = var_f_u

    lam = np.where(np.abs(denom) > 0.0, cov_yf_l / denom, 0.0)
    est = mean_y_l + lam * (mean_f_u - mean_f_l)
    return est


def bound(sd,step,weight,beta):
    return weight * (sd + 2*beta / np.sqrt(step)) / step 

def full_naive_k(b,p,q,n=100,k=1024):
    y = np.random.binomial(1,b,size=(n,k))
    f = np.random.binomial(1, y*p + (1-y) * (1-q) )
    
    return np.mean(y,0)
    

def naive_strat_k(b,p,q,n=100,k=1024):

    w0 = b * (1-p) + (1-b) * q
    w1 = 1 - w0
    
    #Correct label conditional on labels 0,1 
    p0 =  (1-p) * b / w0
    p1 = p * b / w1 
    
    n0 = np.maximum(np.floor(w0*n).astype(int),1)
    n1 = np.maximum(np.floor(w1*n).astype(int),1)
    
    r0 = np.random.binomial(1,p0,size=(n0,k))
    r1 = np.random.binomial(1,p1,size=(n1,k))
    
    m0 = np.mean(r0,0)
    m1 = np.mean(r1,0)
    
    return w0 * m0 + w1 * m1
    
    
def adaptive_strat_k(b,p,q,n=100,k=1024,beta="standard"):
    #beta = 0.1
    if beta == "standard":
        beta = np.sqrt(4.5*np.log(n))
    
    #Approach: Prefill r, then simply move the index 
    #For this, we need to invert the data generating process... 
    
    #probability of predicted labels 0,1 
    w0 = b * (1-p) + (1-b) * q
    w1 = 1 - w0
    
    #Correct label conditional on labels 0,1 
    p0 =  (1-p) * b / w0
    p1 = p * b / w1 
    
    starget0 = w0 * np.sqrt(p0*(1-p0))
    starget1 = w1 * np.sqrt(p1*(1-p1))
    
    r0 = np.random.binomial(1,p0,size=(n,k))
    r1 = np.random.binomial(1,p1,size=(n,k))
    
    
    n0 = np.zeros(k, dtype=int) + 2
    n1 = np.zeros(k, dtype=int) + 2 #Initial two draws per arm
    
    s0 = r0[:2].sum(0)
    s1 = r1[:2].sum(0)  #Initial sums of inputs. These are also the sum of squares because everything is binary! 
    
    idx = np.arange(k)
    
    for _ in range(max(n - 5, 0)):
        std0 = np.sqrt((s0 - (s0)**2 / n0) / (n0-1))
        std1 = np.sqrt((s1 - (s1)**2 / n1) / (n1-1))
        
        
        b0 = bound(std0, n0, w0, beta)
        b1 = bound(std1, n1, w1, beta)

        choose0 = b0 >= b1
        choose1 = ~choose0

        if choose0.any():        
            i0 = idx[choose0]
            s0[i0] += r0[n0[i0],i0]
            n0[i0] += 1

        if choose1.any():
            i1 = idx[choose1]
            s1[i1] += r1[n1[i1],i1]
            n1[i1] += 1
     
    m0 = s0 / n0
    m1 = s1 / n1  
    #print(np.sqrt(p0*(1-p0)),w0,np.sqrt(p1*(1-p1)),w1)
    #print("optimal_a", starget0/(starget0 + starget1), starget1/(starget0+starget1))
    #print("optimal",np.mean(w0 * std0) / (np.mean(w0*std0) + np.mean(w1 * std1)), 
    #      np.mean(w1 * std1) / (np.mean(w0*std0) + np.mean(w1 * std1))  )
    #print("empirical",np.sum(n0)/np.sum(n0+n1),np.sum(n1)/np.sum(n0+n1))
    print("adaptive",n0.mean(),n1.mean(),n0.std(),n1.std())
    return w0 * m0 + w1 * m1


def adaptive_strat_k_alternative(b,p,q,n=100,k=1024,phase_schedule = "quadratic"):
    #Approach: Prefill r, then simply move the index 
    #For this, we need to invert the data generating process... 
    
    #probability of predicted labels 0,1 
    w0 = b * (1-p) + (1-b) * q
    w1 = 1 - w0
    
    print(w0,w1)
    
    #Correct label conditional on labels 0,1 
    p0 =  (1-p) * b / w0
    p1 = p * b / w1 
    
    starget0 = w0 * np.sqrt(p0*(1-p0))
    starget1 = w1 * np.sqrt(p1*(1-p1))
    
    r0 = np.random.binomial(1,p0,size=(n,k))
    r1 = np.random.binomial(1,p1,size=(n,k))
    
    
    # --- Fast range-sum updates via prefix sums (implementation detail) --------------
    # sum r[a:b, j] = pref[b, j] - pref[a, j]
    pref0 = np.vstack([np.zeros((1, k), dtype=int), np.cumsum(r0, axis=0)])
    pref1 = np.vstack([np.zeros((1, k), dtype=int), np.cumsum(r1, axis=0)])
    
    if phase_schedule == "quadratic":
        Ns = [4 * i**2 for i in range(1,1+int(np.sqrt(n/4)))] 
        if not Ns[-1] >= n-1:
            Ns.append(n)
    elif phase_schedule == "linear":
        Ns = [4] + [4+25*i for i in range(int(n/25)-1) ] 
        if not Ns[-1] >= n-1:
            Ns.append(n)
    else:
        assert False, "Phase schedule not defined"
        
    if Ns[0] != 4:
        raise ValueError("This implementation assumes Ns[0] == 4 so we can prefill with 2+2 draws.")
    
    n0 = np.zeros(k, dtype=int) + 2
    n1 = np.zeros(k, dtype=int) + 2 #Initial two draws per arm
    
    s0 = r0[:2].sum(0)
    s1 = r1[:2].sum(0)  #Initial sums of inputs. These are also the sum of squares because everything is binary! 
    
    idx = np.arange(k)
    
    for phase in range(len(Ns)-1): #We already did the first phase of 4 pulls, so there's one less phase
        
        var0 = (s0 - (s0**2) / n0) / (n0 - 1)
        var1 = (s1 - (s1**2) / n1) / (n1 - 1)
        std0 = np.sqrt(np.clip(var0, 0.0, None))
        std1 = np.sqrt(np.clip(var1, 0.0, None))

        
        new_pulls = Ns[phase+1] - Ns[phase] #Number of pulls to allocate in new step 
        
        
        extras = new_pulls - 2 
        
        # ---------------------------------------------------------------------------
        # Step k: "Computation of the allocations M_i^k."
        # We force one draw per stratum, then compute the integer extras \\tilde m_i^k.
        # Procedure b): if all \\hatσ=0 -> do as in a) with m_i^k = p_i * extras.
        # Otherwise: compute continuous m_i^k via points (i),(ii),(iii), then systematic rounding.
        # ---------------------------------------------------------------------------

        if extras == 0:
            # Only the forced 1 draw per stratum (Eq. (1.1))
            tilde0 = np.zeros(k, dtype=int)
            tilde1 = np.zeros(k, dtype=int)

        else:
            # Effective "p_i * \\hatσ_i" terms used repeatedly in the paper formulas.
            # Here p_i are stratum probabilities => w0,w1 in your model.
            B0 = w0 * std0
            B1 = w1 * std1

            all_zero = (B0 == 0) & (B1 == 0)

            # We'll compute continuous m0, m1 (paper's m_i^k in R_+),
            # then apply systematic rounding to get \\tilde m (integers).
            m0 = np.zeros(k, dtype=float)
            m1 = np.zeros(k, dtype=float)

            # ---- Case: all \\hatσ=0 ------------------------------------------------
            # Paper b): "In case \\hatσ_i^{k-1} = 0 for all i, we do as before."
            # "As before" refers to a): set m_i^k = p_i * (Nk-Nk-1-I) before systematic rounding.
            m0[all_zero] = w0 * extras
            m1[all_zero] = w1 * extras  # ensures m0+m1=extras in reals

            # ---- Case: at least one \\hatσ > 0 -------------------------------------
            nontrivial = ~all_zero
            if np.any(nontrivial):

                # Paper b) first says:
                # "For the indexes i such that \\hatσ_i^{k-1} = 0, set m_i^k = 0."
                # With two strata this means:
                only0_pos = nontrivial & (B0 > 0) & (B1 == 0)
                only1_pos = nontrivial & (B1 > 0) & (B0 == 0)
                both_pos  = nontrivial & (B0 > 0) & (B1 > 0)

                # If only one positive: it gets all extras (since the other is fixed at 0)
                m0[only0_pos] = float(extras)
                m1[only0_pos] = 0.0
                m0[only1_pos] = 0.0
                m1[only1_pos] = float(extras)

                # If both positive: apply points (i),(ii),(iii) specialized to I_k=2
                if np.any(both_pos):
                    # ------------------- Point (i) -----------------------------------
                    # Compute A_i = (N_i^{k-1}+1)/(p_i \\hatσ_i^{k-1}) and sort decreasing.
                    # Paper writes: compute (N_i^{k-1}+1)/(p_i \\hatσ_i^{k-1})
                    # and sort in decreasing order; denote ordered quantities by (i).
                    n0b = n0[both_pos].astype(float)
                    n1b = n1[both_pos].astype(float)
                    B0b = B0[both_pos]
                    B1b = B1[both_pos]

                    A0 = (n0b + 1.0) / B0b
                    A1 = (n1b + 1.0) / B1b

                    # order: (1) is the larger A (decreasing order)
                    take0_as_first = A0 >= A1  # boolean per replication in this subset

                    # We'll define "first" and "second" quantities according to this order.
                    B_first  = np.where(take0_as_first, B0b, B1b)
                    B_second = np.where(take0_as_first, B1b, B0b)
                    n_first  = np.where(take0_as_first, n0b, n1b)
                    n_second = np.where(take0_as_first, n1b, n0b)
                    A_first  = np.where(take0_as_first, A0, A1)

                    # ------------------- Point (ii) ----------------------------------
                    # For i=1..I_k compute:
                    #   T_i = (extras + sum_{j=i+1..I_k}(N_{(j)}^{k-1}+1)) / (sum_{j=i+1..I_k} p_{(j)} \\hatσ_{(j)} )
                    #
                    # For I_k=2, only i=1 is meaningful (since for i=2 the denominator is empty/0).
                    # So:
                    #   T_1 = (extras + (n_second+1)) / B_second
                    T1 = (extras + (n_second + 1.0)) / B_second

                    # Define i* as the last i such that:
                    #   (N_{(i)}^{k-1}+1)/(p_{(i)} \\hatσ_{(i)}) >= T_i
                    # For I_k=2:
                    #   if A_first >= T1 then i*=1 else i*=0
                    i_star_is_1 = A_first >= T1

                    # ------------------- Point (iii) ---------------------------------
                    # If i* = 0: both strata are active and:
                    #   m_{(i)} = (p_{(i)} \\hatσ_{(i)}) * T0 - (N_{(i)}^{k-1}+1)
                    # where
                    #   T0 = (extras + (n_first+1) + (n_second+1)) / (B_first + B_second)
                    #
                    # If i* = 1: first gets 0, second gets all extras.
                    T0 = (extras + (n_first + 1.0) + (n_second + 1.0)) / (B_first + B_second)

                    m_first  = np.where(i_star_is_1, 0.0, B_first  * T0 - (n_first  + 1.0))
                    m_second = np.where(i_star_is_1, float(extras), B_second * T0 - (n_second + 1.0))

                    # Numerical safety: clip to feasible set and enforce sum close to extras
                    m_first  = np.clip(m_first,  0.0, float(extras))
                    m_second = np.clip(m_second, 0.0, float(extras))

                    # Map back from (first,second) ordering to original (m0,m1)
                    m0_both = np.where(take0_as_first, m_first,  m_second)
                    m1_both = np.where(take0_as_first, m_second, m_first)

                    m0[both_pos] = m0_both
                    m1[both_pos] = m1_both

            # -----------------------------------------------------------------------
            # "Finally we deduce (\\tilde m_1^k,...,\\tilde m_I^k) in N^I
            # by the systematic sampling procedure described in a)."
            #
            # Paper a) defines:
            #   \\tilde m_i^k = floor(m_1+...+m_i) - floor(m_1+...+m_{i-1})
            # For I=2 this becomes:
            #   \\tilde m_0 = floor(m0)
            #   \\tilde m_1 = extras - floor(m0)   (since m0+m1=extras in reals, up to fp)
            # -----------------------------------------------------------------------
            tilde0 = np.floor(m0 + 1e-12).astype(int)
            tilde0 = np.clip(tilde0, 0, extras)
            tilde1 = extras - tilde0

        # Eq. (1.1): M_i^k = 1 + \\tilde m_i^k
        add_n0 = 1 + tilde0
        add_n1 = 1 + tilde1
        assert np.all(add_n0 + add_n1 == new_pulls)

        # ---------------------------------------------------------------------------
        # "Drawings of the X_i's": draw M_i^k new samples in each stratum,
        # update N_i^k = N_i^{k-1} + M_i^k
        # (In our simulation we already generated r0,r1 and just consume the next rows.)
        # ---------------------------------------------------------------------------
        old_n0 = n0.copy()
        old_n1 = n1.copy()
        n0 = n0 + add_n0
        n1 = n1 + add_n1

        # Update sums s_i by adding the newly consumed block of samples
        s0 = s0 + (pref0[n0, idx] - pref0[old_n0, idx])
        s1 = s1 + (pref1[n1, idx] - pref1[old_n1, idx])
     
    m0 = s0 / n0
    m1 = s1 / n1  
    #for i in range(k):
    #    print(n0[i],n1[i],(B0/(B0+B1))[i],(B1/(B0+B1))[i])
    print("adaptive_alt",n0.mean(),n1.mean(),n0.std(),n1.std())
    return w0 * m0 + w1 * m1


def get_conditional_stds(b=0.9,p=0.9,q=0.9):
    pf1 = b * p + (1-b) * (1-q)
    
    p0 = (1-p)*b/(1-pf1)
    p1 = p * b / pf1 
    return np.sqrt(p0*(1-p0)),np.sqrt(p1*(1-p1))
```

```python
def scaleplot(ns,k=100,b=0.9,p=0.9,q=0.9,ppi=True,strat=True,baseline=True):

    #print(get_conditional_stds(b,p,q))
    y = np.random.binomial(1,b,size=(1000000))
    f = np.random.binomial(1, y*p + (1-y) * (1-q) )
    
    def mse(results):
        return np.mean((np.array(results) - b)**2)
    
    def var(results):
        return np.var(results)

    def bias(results):
        return np.mean(np.array(results) - b)
    
    w0 = b * (1-p) + (1-b) * q
    w1 = 1 - w0
    
    #Correct label conditional on labels 0,1 (For debug)
    p0 =  (1-p) * b / w0
    p1 = p * b / w1 
    
    starget0 = w0 * np.sqrt(p0*(1-p0))
    starget1 = w1 * np.sqrt(p1*(1-p1))
    print("target",starget0/(starget0+starget1),starget1/(starget0+starget1))
    
    if baseline:
        baseline_results = {key: full_naive_k(b,p,q,n=key,k=k) for key in ns}
        mse_baseline = [mse(baseline_results[n])*n for n in ns ]
    
    if ppi:
        ppi_results = {key: ppi_k(b,p,q,n=key,k=k) for key in ns}
        mse_ppi = [mse(ppi_results[n])*n for n in ns ]
        pred_ppi = np.var(y) * (1-np.corrcoef(y,f)[0,1]**2)
        pred_ppi_gpt = w0 * np.var(y[f==0]) + w1*np.var(y[f==1])
        
    if strat:
        strat_results_baseline = {key: naive_strat_k(b,p,q,n=key,k=k) for key in ns}
        strat_results = {key: adaptive_strat_k(b,p,q,n=key,k=k) for key in ns}
        strat_results_b = {key: adaptive_strat_k(b,p,q,n=key,k=k,beta=0.5) for key in ns}
        strat_results_c = {key: adaptive_strat_k_alternative(b,p,q,n=key,k=k,phase_schedule="linear") for key in ns}
        
        mse_strat_baseline = [mse(strat_results_baseline[n])*n for n in ns ]
        mse_strat = [mse(strat_results[n])*n for n in ns ]
        mse_strat_b = [mse(strat_results_b[n])*n for n in ns ]
        mse_strat_c = [mse(strat_results_c[n])*n for n in ns ]
        
        pred_strat = (w0 * np.std(y[f==0]) + w1*np.std(y[f==1]))**2
    

    pred_naive = np.var(y)
    
    legend = []
    
    plt.plot(ns,[pred_naive for n in ns],color="black",linestyle="--",alpha=0.5,label='_nolegend_')

    
    if baseline:
        plt.plot(ns,mse_baseline,color="black")  
        legend.append("Naive sampling")
    
    if ppi:
        plt.plot(ns,[pred_ppi for n in ns],color="blue",linestyle="--",alpha=0.5,label='_nolegend_')
        #plt.scatter(ns,[pred_ppi_gpt for n in ns],color="blue",linestyle="--",alpha=0.5)
        legend.append("PPI (Empirical)")
        plt.plot(ns,mse_ppi,color="blue")
    if strat:
        plt.plot(ns,mse_strat_baseline,color="red")
        legend.append("Basic Strata")
        plt.plot(ns,[pred_strat for n in ns],color="orange",linestyle="--",alpha=0.5,label='_nolegend_')
        plt.plot(ns,mse_strat,color="orange")
        legend.append("Optimized Strata (Robust)")
        plt.plot(ns,mse_strat_b,color="darkorange")
        legend.append("Optimized Strata (Aggressive)")
        plt.plot(ns,mse_strat_c,color="purple")
        legend.append("Optimized Strata (Non-Bandit)")

    print(pred_strat/pred_naive)
        
    
    
    plt.ylabel("n*MSE")
    plt.xlabel("n")
    
    plt.legend(legend)
    
    
    #plt.legend(["ppi","stratified"])
    
    plt.ylim(0.0,None)
    plt.show()
    
    
    
#Check separate scaling
#PPI: Linear in k and n 
#Stratification: Linear-ish in k 
```

```python
def ppi(y,f,n=100):
    mask = np.zeros_like(y,dtype=bool)
    subset = np.random.choice(np.arange(len(y)),size=n,replace=False)
    mask[subset] = 1 
    
    yl = y[mask]
    fl = f[mask]
    fr = f[~mask]
    
    lam = np.cov(yl,fl)[0,1] / ((1 + n / len(fr)) * np.var(fr))
    if np.isinf(lam):
        lam = 0.0
        
    #lam = 0.0 For lambda = 0, things seem to behave as expected...

    return lam * (np.mean(fr) - np.mean(fl)) + np.mean(yl)


def adaptive_strat(n=100, beta = 1 ):
    a0 = y[f==0]
    a1 = y[f==1]
    
    w0 = len(a0)/(len(a0) + len(a1))
    w1 = len(a1)/(len(a0) + len(a1))
    
    r0 = np.zeros(n)
    r1 = np.zeros(n)
    
    p0 = 0 
    p1 = 0
    
    for i in range(2):
        r0[p0] = np.random.choice(a0)
        r1[p1] = np.random.choice(a1)
        p0 += 1
        p1 += 1
        
    while p0+p1 < n-1:
        b0 = bound(np.std(r0[:p0]),p0, w0, beta)
        b1 = bound(np.std(r1[:p0]),p1, w1, beta)
        
        if b0 >= b1:
            r0[p0] = np.random.choice(a0)
            p0 += 1
            
        else:
            r1[p1] = np.random.choice(a1)
            p1+=1
    
    return w0 *  np.mean(r0[:p0]) + w1 * np.mean(r1[:p1])
```

```python
def var_strat(x,y):
    p_total = len(x) + len(y)
    px = len(x) / p_total
    py = len(y) / p_total
    return  2*(px**2 * np.var(x) + py**2 * np.var(y))
```

```python
def var_strat(x,y,w1=0.5,w2=0.5):
    # renormalize w1 and w2...
    w1n = w1 / (w1+w2)
    w2n = w2 / (w1+w2)
    p_total = len(x) + len(y)
    px = len(x) / p_total
    py = len(y) / p_total
    return  (px**2 * np.var(x) / w1n + py**2 * np.var(y) / w2n)
```

```python
    
def substd(r1,n1):
    n, k = r1.shape
    m = n1.max()
    x = r1[:m].astype(float)          # only up to needed max length
    cs  = x.cumsum(0)
    cs2 = (x*x).cumsum(0)

    j = np.arange(k)
    i = n1 - 1
    s  = cs[i,  j]
    s2 = cs2[i, j]

    var = (s2 - s*s/n1) / (n1 - 1)
    std = np.sqrt(np.maximum(var, 0.0))
    return std
```

```python
def adaptive_strat_k(y,f,n=100,k=1024):
    
    beta = np.sqrt(4.5*np.log(n))
    
    # Prefill r 
    
    a0 = y[f == 0]
    a1 = y[f == 1]

    w0 = a0.size / (a0.size + a1.size)
    w1 = 1.0 - w0

    # r0/r1 mirror your original arrays, but vectorized over k
    r0 = np.full((k, n), np.nan, dtype=float)
    r1 = np.full((k, n), np.nan, dtype=float)

    p0 = np.zeros(k, dtype=int)
    p1 = np.zeros(k, dtype=int)

    # Warm start: for i in range(2): sample one from each stratum
    r0[:, 0] = np.random.choice(a0, size=k)
    r1[:, 0] = np.random.choice(a1, size=k)
    r0[:, 1] = np.random.choice(a0, size=k)
    r1[:, 1] = np.random.choice(a1, size=k)
    p0 += 2
    p1 += 2
    
    sum0 = r0[:, 0] + r0[:, 1]
    sum1 = r1[:, 0] + r1[:, 1]
    sumsq0 = r0[:, 0] * r0[:, 0] + r0[:, 1] * r0[:, 1]
    sumsq1 = r1[:, 0] * r1[:, 0] + r1[:, 1] * r1[:, 1]

    # Same stopping logic as: while p0+p1 < n-1
    # Start at 4 total draws; stop at (n-1) => add (n-5) draws.
    idx = np.arange(k)

    for _ in range(max(n - 5, 0)):
        mean0 = sum0 / p0
        mean1 = sum1 / p1

        var0 = (sumsq0 - (sum0 * sum0) / p0) / (p0 - 1)
        var1 = (sumsq1 - (sum1 * sum1) / p1) / (p1 - 1)

        std0 = np.sqrt(np.maximum(var0, 0.0))
        std1 = np.sqrt(np.maximum(var1, 0.0))
        
        std0 =  np.nanstd(r0[:,:_+5],axis=1)
        std1 = np.nanstd(r1[:,:_+5],axis=1)
        
        #print(np.mean(std0),np.mean(std0b))
        #print(np.mean(std1),np.mean(std1b))
        #std0 = np.nanstd(r0, axis=1)  # equivalent to np.std(r0[:p0]) per experiment
        #std1 = np.nanstd(r1, axis=1)  # equivalent to np.std(r1[:p1]) per experiment

        b0 = bound(std0, p0, w0, beta)
        b1 = bound(std1, p1, w1, beta)

        choose0 = b0 >= b1
        choose1 = ~choose0

        if choose0.any():
            i0 = idx[choose0]
            r0[i0, p0[i0]] = np.random.choice(a0, size=i0.size)
            p0[i0] += 1

        if choose1.any():
            i1 = idx[choose1]
            r1[i1, p1[i1]] = np.random.choice(a1, size=i1.size)
            p1[i1] += 1
    #print(np.mean(std0),np.mean(std1),np.mean(np.nanstd(r0,axis=1)),np.mean(np.nanstd(r1,axis=1)))
    print(np.mean(std0),np.mean(std1))
    return w0 * np.nanmean(r0, axis=1) + w1 * np.nanmean(r1, axis=1)
```
