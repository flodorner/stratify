```python
results = []
```

```python
for model in models:
    with open(model) as f:
        d = json.load(f)
    performance = []
    for i in range(1000):
        
        if d["request_states"][i]["result"]["success"] is False:
            performance += [0]
        else:
            score = 0
            try:
                score += int(d["request_states"][i]["annotations"]["omni_math"]["claude_equivalence_judgement"])/3
            except:
                print(model,i)
            try:
                score += int(d["request_states"][i]["annotations"]["omni_math"]["gpt_equivalence_judgement"])/3
            except:
                print(model,i)
            try:
                score += int(d["request_states"][i]["annotations"]["omni_math"]["llama_equivalence_judgement"])/3
            except:
                print(model,i)
            performance += [score]

    results.append(performance)
```
